```python
# 위 데이터로부터 이것도 주말 여부를 구분하는 것이 좋아보이는데요, 사용일자만 가지고 주말인지를 알아내는 법이 따로 필요합니다.
# 모듈러 연산을 이용해 날짜가 주말인지 아닌지를 판별하는 함수를 만들거나 찾아 적용시켜 맨 오른쪽 열에 따로 붙이겠습니다. 

# https://codechacha.com/ko/python-is-today-weekend/ <<- 그 함수임

from datetime import datetime, date
today = datetime(2022, 3, 21)   # 2022-03-21 => Monday = 0
print(today.weekday())
```

```python
def check_weekend(year, month, day):
    weekday = datetime(year, month, day).weekday()
    if weekday <= 4:
        return False # 주말 아님
    if weekday > 4:
        return True # 주말 맞음

for i in range(len(subway_data)):
    if i%1000 == 0: print(i, 'is now on track')
    date = subway_data.loc[i, ['사용일자']][0]
    year = date//10000
    date = date%10000
    month = date//100
    day = date%100

    subway_data.loc[i, ['주말여부']] = check_weekend(year, month, day)
```

```python
i=0
for date in list(subway_coord['사용일자'].unique()):
    is_wkd = subway_coord[subway_coord['사용일자']==date]['주말여부'].unique()[0]
    for station in list(subway_coord['역명'].unique()):
        if i%1000==0: print(i, "is now on track") # 
        get_in_sum = subway_coord[(subway_coord['사용일자']==date) & (subway_coord['역명']==station)]['승차총승객수'].sum()
        get_off_sum = subway_coord[(subway_coord['사용일자']==date) & (subway_coord['역명']==station)]['하차총승객수'].sum()
        latitude = subway_coord[subway_coord['역명']==station]['위도'].unique()[0]
        longitude = subway_coord[subway_coord['역명']==station]['경도'].unique()[0]
        subway_coord_sum.loc[i, :] = [date, station, get_in_sum, get_off_sum, is_wkd, latitude, longitude]
        i+=1

        # 54분 18초 걸림,,,
```

```python
# 위에서 전처리 후 내보내기 했던 파일을 다시 가져옵니다.
data_coord = pd.read_excel("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/지하철/서울 지하철 주소 지오코딩용.xlsx")

# 2016년 1월~12월

subway_data = pd.read_csv("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/지하철/CARD_SUBWAY_MONTH_201601.csv", encoding ="cp949")
for month in range(2,13):
    if month < 10:
        df = pd.read_csv("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/지하철/CARD_SUBWAY_MONTH_20160"+str(month)+".csv", encoding ="cp949")
    else:
        df = pd.read_csv("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/지하철/CARD_SUBWAY_MONTH_2016"+str(month)+".csv", encoding ="cp949")
    subway_data = pd.concat([subway_data, df], axis=0) # 열 방향으로 합치기

subway_data = subway_data.reset_index(drop=True)

# 역명 표기 통일 (괄호 이하 삭제)
for i in range(len(subway_data)):
    subway_data.loc[i, '역명'] = subway_data.loc[i, '역명'].split("(")[0]  

for i in range(len(data_coord)):
    data_coord.loc[i, '역명'] = data_coord.loc[i, '역명'].split("(")[0]  


from datetime import datetime, date
def check_weekend(year, month, day):
    weekday = datetime(year, month, day).weekday()
    if weekday <= 4:
        return False # 주말 아님
    if weekday > 4:
        return True # 주말 맞음

for i in range(len(subway_data)):
    if i%1000 == 0: print(i, 'is now on track: 주말 여부 검사')
    date = subway_data.loc[i, ['사용일자']][0]
    year = date//10000
    date = date%10000
    month = date//100
    day = date%100

    subway_data.loc[i, ['주말여부']] = check_weekend(year, month, day)

data_coord_2 = data_coord[['역명', '위도', '경도']].drop_duplicates('역명') 

subway_coord = pd.merge(left = subway_data, right = data_coord_2, how = "inner", on = "역명")

subway_coord_sum = pd.DataFrame(columns=['사용일자', '역명', '승차총승객수', '하차총승객수', '주말여부', '위도', '경도'])

i=0
for date in list(subway_coord['사용일자'].unique()):
    is_wkd = subway_coord[subway_coord['사용일자']==date]['주말여부'].unique()[0]
    for station in list(subway_coord['역명'].unique()):
        if i%1000==0: print(i, "is now on track : 다른 호선 같은 역 이용객 수 합치기") 
        get_in_sum = subway_coord[(subway_coord['사용일자']==date) & (subway_coord['역명']==station)]['승차총승객수'].sum()
        get_off_sum = subway_coord[(subway_coord['사용일자']==date) & (subway_coord['역명']==station)]['하차총승객수'].sum()
        latitude = subway_coord[subway_coord['역명']==station]['위도'].unique()[0]
        longitude = subway_coord[subway_coord['역명']==station]['경도'].unique()[0]
        subway_coord_sum.loc[i, :] = [date, station, get_in_sum, get_off_sum, is_wkd, latitude, longitude]
        i+=1

wkd_subway_data = subway_coord_sum[subway_coord_sum['주말여부']==True].reset_index(drop=True)
not_wkd_subway_data = subway_coord_sum[subway_coord_sum['주말여부']==False].reset_index(drop=True)

wkd_subway_mean_data = pd.DataFrame(columns=['역명', '승차총승객수평균', '하차총승객수평균', '주말여부', '위도', '경도'])
for i, station in enumerate(list(wkd_subway_data['역명'].unique())):
    print(i, "is now on working : 각 역마다 이용객 수 평균값 내기(주말)")
    get_on_mean = wkd_subway_data.loc[wkd_subway_data["역명"]==station]['승차총승객수'].mean()
    get_off_mean = wkd_subway_data.loc[wkd_subway_data["역명"]==station]['하차총승객수'].mean()
    latitude = wkd_subway_data.loc[wkd_subway_data["역명"]==station]['위도'].unique()[0]
    longitude = wkd_subway_data.loc[wkd_subway_data["역명"]==station]['경도'].unique()[0]
    wkd_subway_mean_data.loc[i, :] = [station, get_on_mean, get_off_mean, True, latitude, longitude]

not_wkd_subway_mean_data = pd.DataFrame(columns=['역명', '승차총승객수평균', '하차총승객수평균', '주말여부', '위도', '경도'])
for i, station in enumerate(list(not_wkd_subway_data['역명'].unique())):
    print(i, "is now on working : 각 역마다 이용객 수 평균값 내기(평일)")
    get_on_mean = not_wkd_subway_data.loc[not_wkd_subway_data["역명"]==station]['승차총승객수'].mean()
    get_off_mean = not_wkd_subway_data.loc[not_wkd_subway_data["역명"]==station]['하차총승객수'].mean()
    latitude = not_wkd_subway_data.loc[not_wkd_subway_data["역명"]==station]['위도'].unique()[0]
    longitude = not_wkd_subway_data.loc[not_wkd_subway_data["역명"]==station]['경도'].unique()[0]
    not_wkd_subway_mean_data.loc[i, :] = [station, get_on_mean, get_off_mean, False, latitude, longitude]    

not_wkd_subway_mean_data.to_csv("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/지하철/2016년 서울 지하철 평일 하루당 총승차 총하차 수 평균.csv")  
wkd_subway_mean_data.to_csv("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/지하철/2016년 서울 지하철 주말 하루당 총승차 총하차 수 평균.csv")  
```
