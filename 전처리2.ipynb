```python
import pandas as pd
import os
# 로드할 파일의 경로
dir = '/content/drive/MyDrive/기쁨코인/dataset/X_attributes'
for folder in os.listdir(dir):
  print(folder)
  for file in os.listdir(dir+'/'+folder):
    data = pd.read_csv(dir+'/'+folder+'/'+file)
    new_data = pd.DataFrame(data[['좌표정보(X)', '좌표정보(Y)', '지번주소']])

    nan = new_data[new_data['좌표정보(X)'].isnull()]

    name = os.path.basename(file)[:-4]
    nan['지번주소'].to_excel(f'/content/drive/MyDrive/기쁨코인/dataset/nan_{name}.xls')
```
