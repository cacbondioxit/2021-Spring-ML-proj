```python
path = "/content/drive/MyDrive/기쁨코인/dataset/table/table_2016.csv"
table_2016 = pd.read_csv(path).drop("Unnamed: 0", axis=1) 
df1 = pd.read_excel("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/인구밀도/2016 행정동별 인구밀도.xls")
pop_density = df1[df1['동']=='소계']
pop_density = pop_density.drop(['기간', '동'], axis=1).reset_index(drop=True)
pop_density
pop_num = pd.read_excel("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/인구/2015-2017 서울시 자치구별 주민등록인구.xls")
pop_num = pop_num.loc[pop_num['기간']==2016].reset_index(drop=True).drop(0, axis=0).drop(['기간', '계'], axis=1)
pop = pop_density.merge(pop_num, how='inner')
df2 = pd.read_excel("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/지역내 총생산/2016 지역내총생산.xls").drop("기간", axis=1)
df3 = pd.read_excel("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/자치구별 총 부가가치/2016 자치구별 총부가가치.xls")
df3 = df3.drop(['기간', '광업 ', '광업 증감률'], axis=1) # 결측치가 존재했었는데 NaN이 아니라 -, ... 이런식이었음
for i in range(len(table_2016)):
    table_2016.loc[i, '자치구'] = table_2016['address'][i].split()[1]
table_2016_1 = table_2016.merge(pop, on=['자치구'], how='inner')
table_2016_2 = table_2016_1.merge(df2, on=['자치구'], how='inner')
table_2016_3 = table_2016_2.merge(df3, on=['자치구'], how='inner')
table_2016_3.to_csv("/content/drive/MyDrive/기쁨코인/dataset/table/TABLE_2016.csv")
```
