```python
import numpy as np
import pandas as pd
```

```python
# 참고로 이 코드는 15 16 17 평균값을 따로 계산하여, 주말도 따로 떼어서 총 6개의 데이터 프레임을 만들고 있습니다. 
# 3개년도 총 평균을 계산해서 주말과 평일 둘 다를 하나의 데이터프레임에 집어넣는 코드도 아래에 있어요.

year_list = "2015", "2016", "2017"
name_list = df_weekend['버스정류장ARS번호'].unique()

for year in year_list:
    df = pd.read_csv('/content/drive/MyDrive/기쁨코인/dataset/버스붕붕.zip (Unzipped Files)/주중-주말-승하차객/BUS_filte_'+year+'.csv', 
                    names=['주차', '버스정류장ARS번호', '역명', '승차총승객수', '하차총승객수', 'latitude', 'longitude', '주말 여부']
                    )
    df = df.drop(columns = ['역명']) # 버스 정류장 이름에 중복된 속성이 많아서 쓸 수 없음. 예를들어 "소방소앞" 정류장은 4개나 있어요. 따라서 버립니다.

    # 승하차 인원은 주말 여부에 따라 다를 것 같아서 따로 취급합니다.

    # 주말
    df_weekend = df[df['주말 여부']==True]
    df_weekend_mean = pd.DataFrame(columns=['버스정류장ARS번호', '승차총승객수', '하차총승객수', 'longitude', 'latitude']) # 평균값을 기록하기 위한 데이터 프레임
    ARS_list = df_weekend['버스정류장ARS번호'].unique() # ARS 코드 리스트

    for i, ARS in enumerate(name_list): 
        # 평균값 기록
        
        get_in = df_weekend[df_weekend['버스정류장ARS번호']==ARS]['승차총승객수'].mean() # 일주일 간 승차 총 승객수 평균값
        get_off = df_weekend[df_weekend['버스정류장ARS번호']==ARS]['하차총승객수'].mean() # 일주일 간 하차 총 승객수 평균값
        longitude = np.float64(df_weekend[df_weekend['버스정류장ARS번호']==ARS_list[i]]['longitude'].unique()) # 버스 정류장의 longitude
        latitude = np.float64(df_weekend[df_weekend['버스정류장ARS번호']==ARS_list[i]]['latitude'].unique()) # 버스 정류장의 latitude
        df_weekend_mean.loc[i, ['버스정류장ARS번호', '승차총승객수', '하차총승객수', 'longitude', 'latitude']] = ARS, get_in, get_off, longitude, latitude
        print(i, ARS, "has been done")

    # 평일
    df_not_weekend = df[df['주말 여부']==False]
    df_not_weekend_mean = pd.DataFrame(columns=['버스정류장ARS번호', '승차총승객수', '하차총승객수', 'longitude', 'latitude'])
    ARS_list = df_not_weekend['버스정류장ARS번호'].unique()

    for i, ARS in enumerate(name_list):
        # 평균값 기록

        get_in = df_not_weekend[df_not_weekend['버스정류장ARS번호']==ARS]['승차총승객수'].mean()
        get_off = df_not_weekend[df_not_weekend['버스정류장ARS번호']==ARS]['하차총승객수'].mean()
        longitude = np.float64(df_not_weekend[df_not_weekend['버스정류장ARS번호']==ARS_list[i]]['longitude'].unique())
        latitude = np.float64(df_not_weekend[df_not_weekend['버스정류장ARS번호']==ARS_list[i]]['latitude'].unique())
        df_not_weekend_mean.loc[i, ['버스정류장ARS번호', '승차총승객수', '하차총승객수', 'longitude', 'latitude']] = ARS, get_in, get_off, longitude, latitude
        print(i, ARS, "has been done")

    df_weekend_mean.to_csv("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/BUS/평일버스승하차인원_"+year+".csv") # csv로 내보내기
    df_not_weekend_mean.to_csv("/content/drive/MyDrive/기쁨코인/dataset/X_attributes_final/BUS/주말버스승하차인원_"+year+".csv")
```

```python
df = df_2015
```

```python
# 평일과 주말을 같이 데이터프레임에 넣으려면 평일과 주말 사이에 ARS_list가 동일해야만 합니다. 그래서 확인해보니 동일하지 않았습니다. 
# 2015년을 보니 ARS가 {18688}인 경우는 평일만 정차합니다.
# 2017년을 보니 ARS가 {1009, 1010, 1011, 1012, 1013, 1014, 1015, 1016, 1017, 1018, 1019, 1020, 1021, 1022, 1023}인 버스정류장은 주말에만 정차합니다.
# 아래 코드에서 A-B가 {1009, 1010, 1011, 1012, 1013, 1014, 1015, 1016, 1017, 1018, 1019, 1020, 1021, 1022, 1023}입니다.
# B-A는 공집합입니다.

ARS_list_1 = df[df['주말 여부']==True]['버스정류장ARS번호'].unique()
ARS_list_2 = df[df['주말 여부']==False]['버스정류장ARS번호'].unique()

A = set(ARS_list_1)  
B = set(ARS_list_2)

print(A-B)
print(B-A)
```

```python
df[df['버스정류장ARS번호']==18688]
# 주차를 보니 2015년 중간에 새로 생긴 것 같습니다. 
```

```python
#주말에만 버스가 정차하는 버스 정류장에 대해
#우리는 "주말여부==False로 해놓고, 총 승객수를 0, 나머지 정보는 그대로 갖는 행을 추가하겠습니다. 
# 또한 우리는 지금 각 주마다의 평균값만 계산할 것이기 때문에, 승객수=0인 행에 대해 1주~52주까지 모든 주차를 넣을 필요는 없습니다.

wkd = list(A-B)

for ARS in wkd:
    idx = df[df['버스정류장ARS번호']==ARS].index[0]
    name = df[df['버스정류장ARS번호']==ARS].loc[idx, '역명']
    latitude = df[df['버스정류장ARS번호']==ARS].loc[idx, 'latitude']
    longitude = df[df['버스정류장ARS번호']==ARS].loc[idx, 'longitude']
    df.loc[len(df)] = [0, ARS, name, 0, 0, latitude, longitude, False]

df.tail(20)
```

```python
# 이제 다시 주말과 평일로 나눕니다.

# 주말
df_weekend = df[df['주말 여부']==True]
ARS_list_1 = df_weekend['버스정류장ARS번호'].unique() # ARS 코드 리스트

# 평일
df_not_weekend = df[df['주말 여부']==False]
ARS_list_2 = df_not_weekend['버스정류장ARS번호'].unique() # ARS 코드 리스트

# 평균값을 기록하기 위한 데이터 프레임
df_mean = pd.DataFrame(columns=['버스정류장ARS번호', '평일 승차 수 평균', '평일 하차 수 평균', '주말 승차 수 평균', '주말 하차 수 평균', 'longitude', 'latitude']) 
```
