```python
import pandas as pd
import os

def data_cleansing_1(name):
    # 로드할 파일의 경로
    path ='/content/drive/MyDrive/기쁨코인/dataset/서울시 '+name+' 인허가 정보.csv'
    data = pd.read_csv(path, encoding='cp949')

    # 모든 테이블이 똑같이 생겼음을 이용해서 2015, 2016, 2017년에 존재했던 가게들을 남게끔 처리합니다
    data['인허가일자1'] = data['인허가일자']/ 10000
    data['폐업일자1'] = data['폐업일자']/ 10000

    data['폐업일자1'] = data['폐업일자1'].fillna(3000)

    data15 = data.loc[data['인허가일자1'] < 2016]
    data16 = data.loc[data['인허가일자1'] < 2017]
    data17 = data.loc[data['인허가일자1'] < 2018]

    data15 = data15.drop('인허가일자1', axis = 1)
    data16 = data16.drop('인허가일자1', axis = 1)
    data17 = data17.drop('인허가일자1', axis = 1)

    data15 = data15.loc[data['폐업일자1'] > 2015]
    data16 = data16.loc[data['폐업일자1'] > 2016]
    data17 = data17.loc[data['폐업일자1'] > 2017]

    # 저장할 csv 파일 이름 및 경로 설정
    os.mkdir('/content/drive/MyDrive/기쁨코인/dataset/'+name)
    data15.to_csv('/content/drive/MyDrive/기쁨코인/dataset/'+name+'/'+name+'_2015.csv', encoding='utf-8')
    data16.to_csv('/content/drive/MyDrive/기쁨코인/dataset/'+name+'/'+name+'_2016.csv', encoding='utf-8')
    data17.to_csv('/content/drive/MyDrive/기쁨코인/dataset/'+name+'/'+name+'_2017.csv', encoding='utf-8')

data_cleansing_1('당구장업')
data_cleansing_1('골프연습장업')
data_cleansing_1('인터넷컴퓨터게임시설제공업')
data_cleansing_1('일반게임제공업')
data_cleansing_1('유원시설업(기타)')
```
